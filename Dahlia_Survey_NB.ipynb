```python
def display_all_surveys():
    data = ref.get()

    if data:
        print("\n--- Survey Records ---")
        for key, value in data.items():
            print(f"\nID: {key}")
            for k, v in value.items():
                print(f"{k}: {v}")
    else:
        print("No survey data found.")
```

```python
while True:
    print("\n--- Firebase Survey Database Menu ---")
    print("1. Display All Surveys")
    print("2. Add New Survey")
    print("3. Update Survey")
    print("4. Delete Survey")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    # DISPLAY
    if choice == "1":
        raw_surveys_data = ref.get()
        print("\nSurvey List:")

        surveys_to_display = {}
        if isinstance(raw_surveys_data, dict):
            surveys_to_display = raw_surveys_data
        elif isinstance(raw_surveys_data, list): # Handle cases where Firebase returns a list
            for item in raw_surveys_data:
                if item and 'id' in item: # Ensure item is not None and has an 'id' key
                    surveys_to_display[str(item['id'])] = item
        # If raw_surveys_data is None, surveys_to_display remains empty

        if surveys_to_display:
            for key, survey in surveys_to_display.items():
                print(
                    f"ID: {survey['id']}, "
                    f"Age: {survey['age']}, "
                    f"Gender: {survey['gender']}, "
                    f"Grade: {survey['grade_level']}, "
                    f"Section: {survey['section']}, "
                    f"Q1: {survey['q1_satisfaction']}, "
                    f"Q2: {survey['q2_usability']}, "
                    f"Q3: {survey['q3_design']}, "
                    f"Q4: {survey['q4_clarity']}, "
                    f"Q5: {survey['q5_recommend']}, "
                    f"Q6: {survey['q6_overall']}, "
                    f"Q7: {survey['q7_comments']}"
                )
        else:
            print("No survey data found.")

    # ADD
    elif choice == "2":
        sid = input("Enter survey ID: ")

        age = int(input("Enter age: "))
        gender = input("Enter gender: ")
        grade_level = input("Enter grade level: ")
        section = input("Enter section: ")

        q1 = int(input("Enter Q1 satisfaction (1-5): "))
        q2 = int(input("Enter Q2 usability (1-5): "))
        q3 = int(input("Enter Q3 design (1-5): "))
        q4 = int(input("Enter Q4 clarity (1-5): "))
        q5 = int(input("Enter Q5 recommend (1-5): "))
        q6 = int(input("Enter Q6 overall (1-5): "))
        q7 = input("Enter Q7 comments: ")

        survey = {
            "id": int(sid),
            "age": age,
            "gender": gender,
            "grade_level": grade_level,
            "section": section,
            "q1_satisfaction": q1,
            "q2_usability": q2,
            "q3_design": q3,
            "q4_clarity": q4,
            "q5_recommend": q5,
            "q6_overall": q6,
            "q7_comments": q7
        }

        ref.child(sid).set(survey)
        print("Survey added successfully!")

    # UPDATE
    elif choice == "3":
        sid = input("Enter ID of survey to update: ")
        survey_ref = ref.child(sid)

        survey = survey_ref.get()

        if survey:
            age = int(input("Enter new age: "))
            gender = input("Enter new gender: ")
            grade_level = input("Enter new grade level: ")
            section = input("Enter new section: ")

            q1 = int(input("Enter new Q1 satisfaction (1-5): "))
            q2 = int(input("Enter new Q2 usability (1-5): "))
            q3 = int(input("Enter new Q3 design (1-5): "))
            q4 = int(input("Enter new Q4 clarity (1-5): "))
            q5 = int(input("Enter new Q5 recommend (1-5): "))
            q6 = int(input("Enter new Q6 overall (1-5): "))
            q7 = input("Enter new Q7 comments: ")

            survey_ref.update({
                "\age": age,
                "gender": gender,
                "grade_level": grade_level,
                "section": section,
                "q1_satisfaction": q1,
                "q2_usability": q2,
                "q3_design": q3,
                "q4_clarity": q4,
                "q5_recommend": q5,
                "q6_overall": q6,
                "q7_comments": q7
            })

            print("Survey updated successfully!")
        else:
            print("Survey not found.")

    elif choice == "4":
        sid = input("Enter ID to delete: ")

        if ref.child(sid).get():
            ref.child(sid).delete()
            print("Survey deleted successfully!")
        else:
            print("Survey not found.")

    elif choice == "5":
        print("Exiting program. Goodbye!")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```

```python
import requests
import json

url = "https://raw.githubusercontent.com/JensineArissaOArabis/Final-Project-CS2/refs/heads/main/Project_Template_ComputerScience2-main/data/survey.json"
response = requests.get(url)
surveys = response.json()
print("JSON file loaded")
ref = db.reference("surveys")

for survey in surveys:
  if 'respondent_id' in survey:
      survey['id'] = survey.pop('respondent_id')
  ref.child(str(survey["id"])).set(survey)
print("Data uploaded successfully!")
```

```python
def display_all_surveys():
    data = ref.get()

    if data:
        print("\n--- Survey Records ---")
        for key, value in data.items():
            print(f"\nID: {key}")
            for k, v in value.items():
                print(f"{k}: {v}")
    else:
        print("No survey data found.")
```

```python
import datetime

while True:
    print("\n--- Firebase Survey Database Menu ---")
    print("1. Display All Surveys")
    print("2. Add New Survey")
    print("3. Update Survey")
    print("4. Delete Survey")
    print("5. See Features")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    match choice:
        case "1":
            raw_surveys_data = ref.get()
            print("\nSurvey List:")

            surveys_to_display = {}
            if isinstance(raw_surveys_data, dict):
                surveys_to_display = raw_surveys_data
            elif isinstance(raw_surveys_data, list): # Handle cases where Firebase returns a list
                for item in raw_surveys_data:
                    if item and 'id' in item: # Ensure item is not None and has an 'id' key
                        surveys_to_display[str(item['id'])] = item
            # If raw_surveys_data is None, surveys_to_display remains empty

            if surveys_to_display:
                for key, survey in surveys_to_display.items():
                    print(
                        f"ID: {survey['id']}, "
                        f"Age: {survey['age']}, "
                        f"Gender: {survey['gender']}, "
                        f"Grade: {survey['grade_level']}, "
                        f"Section: {survey['section']}, "
                        f"Q1: {survey['q1_satisfaction']}, "
                        f"Q2: {survey['q2_usability']}, "
                        f"Q3: {survey['q3_design']}, "
                        f"Q4: {survey['q4_clarity']}, "
                        f"Q5: {survey['q5_recommend']}, "
                        f"Q6: {survey['q6_overall']}, "
                        f"Q7: {survey['q7_comments']}"
                    )
            else:
                print("No survey data found.")

        case "2":
            sid = input("Enter survey ID (number): ")

            age = int(input("Enter age (number): "))
            gender = input("Enter gender (text): ")
            grade_level = input("Enter grade level (text): ")
            section = input("Enter section (text): ")

            q1 = int(input("Enter Q1 satisfaction (number, 1-5): "))
            q2 = int(input("Enter Q2 usability (number, 1-5): "))
            q3 = int(input("Enter Q3 design (number, 1-5): "))
            q4 = int(input("Enter Q4 clarity (number, 1-5): "))
            q5 = int(input("Enter Q5 recommend (number, 1-5): "))
            q6 = int(input("Enter Q6 overall (number, 1-5): "))
            q7 = input("Enter Q7 comments (text): ")

            survey = {
                "id": int(sid),
                "age": age,
                "gender": gender,
                "grade_level": grade_level,
                "section": section,
                "q1_satisfaction": q1,
                "q2_usability": q2,
                "q3_design": q3,
                "q4_clarity": q4,
                "q5_recommend": q5,
                "q6_overall": q6,
                "q7_comments": q7
            }

            ref.child(sid).set(survey)
            print("Survey added successfully!")

        case "3":
            sid = input("Enter ID of survey to update (number): ")
            survey_ref = ref.child(sid)

            survey = survey_ref.get()

            if survey:
                age = int(input("Enter new age (number): "))
                gender = input("Enter new gender (text): ")
                grade_level = input("Enter new grade level (text): ")
                section = input("Enter new section (text): ")

                q1 = int(input("Enter new Q1 satisfaction (number, 1-5): "))
                q2 = int(input("Enter new Q2 usability (number, 1-5): "))
                q3 = int(input("Enter new Q3 design (number, 1-5): "))
                q4 = int(input("Enter new Q4 clarity (number, 1-5): "))
                q5 = int(input("Enter new Q5 recommend (number, 1-5): "))
                q6 = int(input("Enter new Q6 overall (number, 1-5): "))
                q7 = input("Enter new Q7 comments (text): ")

                survey_ref.update({
                    "age": age,
                    "gender": gender,
                    "grade_level": grade_level,
                    "section": section,
                    "q1_satisfaction": q1,
                    "q2_usability": q2,
                    "q3_design": q3,
                    "q4_clarity": q4,
                    "q5_recommend": q5,
                    "q6_overall": q6,
                    "q7_comments": q7
                })

                print("Survey updated successfully!")
            else:
                print("Survey not found.")

        case "4":
            sid = input("Enter ID to delete (number): ")

            if ref.child(sid).get():
                ref.child(sid).delete()
                print("Survey deleted successfully!")
            else:
                print("Survey not found.")

        case "5": # New: See Features
            while True:
                print("\n--- Survey Features ---")
                print("1. Restrict maximum word count for comments")
                print("2. Says what kind of answer is expected (a word, number, date, etc.)")
                print("3. Planned Feature #3")
                print("4. Planned Feature #4")
                print("5. Planned Feature #5")
                print("6. Back to Main Menu")

                feature_choice = input("Enter your choice (1-6): ")

                match feature_choice:
                    case "1":
                        print("\nFeature 1: Restrict maximum word count for comments")
                        maxwords = 250
                        while True:
                            user_input = input("Enter your answer (max 250 words): ")
                            if len(user_input.split()) <= maxwords:
                                break
                            else:
                                print("Too many, (", len(user_input.split()),"words )", ", please make sure the amount of words entered is below 250 and try again")
                        print("Accepted", len(user_input.split()), "word/s entered")
                    case "2":
                        print("\nFeature 2: Says what kind of answer is expected (a word, number, date, etc.)")
                        print("--- Survey Questions and Expected Types ---")

                        sample_data = ref.get()
                        if sample_data:
                            sample_survey = None
                            if isinstance(sample_data, dict):
                                # If it's a dict, get the first entry
                                first_key = next(iter(sample_data))
                                sample_survey = sample_data[first_key]
                            elif isinstance(sample_data, list):
                                # If it's a list, find the first non-None dictionary entry
                                for item in sample_data:
                                    if item is not None and isinstance(item, dict):
                                        sample_survey = item
                                        break

                            if sample_survey: # Only proceed if a valid sample_survey was found
                                for key, value in sample_survey.items():
                                    if key == 'id':
                                        print(f"- ID (number)")
                                    elif key == 'age':
                                        print(f"- Age (number)")
                                    elif key == 'gender':
                                        print(f"- Gender (text)")
                                    elif key == 'grade_level':
                                        print(f"- Grade Level (text)")
                                    elif key == 'section':
                                        print(f"- Section (text)")
                                    elif key.startswith('q') and key.endswith('_satisfaction'):
                                        print(f"- {key.replace('q', 'Q').replace('_satisfaction', ' satisfaction')} (number, 1-5)")
                                    elif key.startswith('q') and key.endswith('_usability'):
                                        print(f"- {key.replace('q', 'Q').replace('_usability', ' usability')} (number, 1-5)")
                                    elif key.startswith('q') and key.endswith('_design'):
                                        print(f"- {key.replace('q', 'Q').replace('_design', ' design')} (number, 1-5)")
                                    elif key.startswith('q') and key.endswith('_clarity'):
                                        print(f"- {key.replace('q', 'Q').replace('_clarity', ' clarity')} (number, 1-5)")
                                    elif key.startswith('q') and key.endswith('_recommend'):
                                        print(f"- {key.replace('q', 'Q').replace('_recommend', ' recommend')} (number, 1-5)")
                                    elif key.startswith('q') and key.endswith('_overall'):
                                        print(f"- {key.replace('q', 'Q').replace('_overall', ' overall')} (number, 1-5)")
                                    elif key == 'q7_comments':
                                        print(f"- Q7 Comments (text)")
                                    else:
                                        print(f"- {key} (unknown type)") # Fallback for unexpected keys
                            else:
                                print("No valid survey data available to display questions.")
                        else:
                            print("No survey data available to display questions.")

                    case "3":
                        print("\nFeature 3: Planned Feature #3")
                    case "4":
                        print("\nFeature 4: Planned Feature #4")
                    case "5":
                        print("\nFeature 5: Planned Feature #5")
                    case "6":
                        print("Returning to Main Menu.")
                        break # Exit the features loop to go back to the main menu
                    case _:
                        print("Invalid choice. Please enter a number between 1 and 6.")

        case "6":
            print("Exiting program. Goodbye!")
            break

        case _:
            print("Invalid choice. Please enter a number between 1 and 6.")
```
